```python
import pandas as pd
import numpy as np
```

## Load Data

```python
df = pd.read_csv('data/HW2-Fall2025-Data.csv')
df
```

```python
# drop first 2 rows and start from index 1
df = df.drop(index=[0, 1])
df = df.reset_index(drop=True)
df.index += 1
df
```

```python
# drop Student ID since studentID = index + 1
df = df.drop(columns=['Student ID'])
df.head()
```

```python
# convert all columns to float
for col in df.columns:
    df[col] = df[col].astype(float)
df.dtypes
```

```python
df.describe()
```

## Covariance Matrix

```python
# get the covariance matrix of the normalized data
cov_matrix = df.cov()
cov_matrix
```

## Eigenvectos and eigenvalues

```python
# find eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print("Eigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")    
print(eigenvectors)
```

## Projections

```python
# first two eigenvectors
W1 = eigenvectors[0]
W2 = eigenvectors[1]
W12 = np.column_stack((W1, W2))
W12
```

```python
# project data onto the first two principal components
pca12 = df.dot(W12)
pca12 = pd.DataFrame(pca12)
pca12
```
